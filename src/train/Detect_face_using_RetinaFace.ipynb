```python
from retinaface import RetinaFace
import cv2
from google.colab.patches import cv2_imshow
```

## Test for single image

```python
image_path = '/content/box.jpg'
```

```python
res = RetinaFace.detect_faces(image_path)
```

```python
res
```

```python
import random
```

```python
image = cv2.imread(image_path)
for (key, value) in res.items():
  xmin, ymin, xmax, ymax = value['facial_area']

  color = (random.randint(0, 200), random.randint(0, 200), random.randint(0, 200))

  cv2.rectangle(image, (xmin, ymin), (xmax, ymax), color, 2)
```

```python
cv2_imshow(image)
```

## Test with video

```python
video_path = '/content/drive/MyDrive/emotion-recognition/assets/test_video.mp4'
video = cv2.VideoCapture(video_path)
video.isOpened()
```

```python
video_codec = cv2.VideoWriter_fourcc(*'mp4v')
output = cv2.VideoWriter('./saved_video.mp4', fourcc=int(video_codec), fps=30, frameSize=(1080, 720))

no_frames = 0
while True:
  ret, frame = video.read()
  if not ret:
    break

  frame = cv2.resize(frame, (1080, 720))
  if no_frames % 5 == 0:
    results = RetinaFace.detect_faces(frame)

    print(f'Frame {no_frames} has : {len(results)} faces.')
    for (_key, value) in results.items():
      xmin, ymin, xmax, ymax = value['facial_area']

      cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (255, 0, 143), 2)

  output.write(frame)
  no_frames += 1

video.release()
output.release()
```
